```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import GridSearchCV
 
```

```python
data = pd.read_csv('../../session_18/code/creditcard.csv')
data.head()
```

```python
data = data.drop(['Time'], axis=1)
data.head()
```

```python
x = data.drop(['Class'], axis=1)
y = data['Class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score , confusion_matrix, classification_report
```

```python
hyper_param = {'criterion':['entropy'], 'max_depth':[4,6], 'n_estimators':[10, 20, 25], 'max_features':['sqrt']}
```

```python
tree_GS_param = GridSearchCV(RandomForestClassifier(),hyper_param, scoring='roc_auc', cv=5, n_jobs=-1)
tree_training = tree_GS_param.fit(x_train, y_train)
print('the best parameter is : ', tree_training.best_params_)
print('the best estimator is : ', tree_training.best_estimator_)
print('the best score is : ', tree_training.best_score_)
```

```python
y_pred = tree_GS_param.predict(x_test)
```

```python
cm1 = confusion_matrix(y_test, y_pred)

print(cm1)

print(classification_report(y_test, y_pred, target_names=["Safe", "Fraud"]))
 
#Calculate sensitivity and specificity
 
total1=sum(sum(cm1))

accuracy1=(cm1[0,0]+cm1[1,1])/total1

print ('Accuracy : ', accuracy1)
 
sensitivity1 = cm1[0,0]/(cm1[0,0]+cm1[0,1])

print('Sensitivity : ', sensitivity1 )
 
specificity1 = cm1[1,1]/(cm1[1,0]+cm1[1,1])

print('Specificity : ', specificity1)
 
```

```python
from sklearn.ensemble import GradientBoostingClassifier, GradientBoostingRegressor

gbc_param = {
    'n_estimators':[100,200],
    'learning_rate':[.01, 0.1],
    'max_depth':[3, 4],
    'subsample':[0.8, 1.0],
    'min_samples_split':[2, 5]
}

tree_GSB_param = GridSearchCV(GradientBoostingClassifier(),gbc_param, scoring='roc_auc', cv=5, n_jobs=-1)
tree_training_bost = tree_GSB_param.fit(x_train, y_train)س
print('the best parameter is : ', tree_training_bost.best_params_)
print('the best estimator is : ', tree_training_bost.best_estimator_)
print('the best score is : ', tree_training_bost.best_score_)
```
