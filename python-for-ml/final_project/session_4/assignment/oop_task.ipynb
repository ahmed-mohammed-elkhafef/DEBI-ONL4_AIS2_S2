```python
class Fibanicchi:
    """
    class Fibanicchi: handles the operation of calculation of fibanicchi series

    Attributes:
            num(int): number that define the length of series
            x(int): first number in series 0
            y(int): second number in series 1 
    """
    def __init__(self, num, x=0, y=1):
        self.num=num
        self.x=x
        self.y=y

    def fibanicchi(self):
        """
        this function calculates fibonacchi series 
        
        param1: num is the length of series,
        type: integer

        param2: frist value in series,
        type: integer

        param3: second value in series,
        type: integer

        return: fibonacchi series,
        rtype: str
        """
        if self.num == 0:
            return "" 
        return f"{self.x}  {Fibanicchi(self.num-1, self.y, self.x+self.y).fibanicchi()}"

f1=Fibanicchi(10)
print(f1.fibanicchi())
```

```python
class Factorial:
    """
    class Factorial: handles the operation of calculation of factorial of a number

    Attributes:
            num(int): number that we want to calculate its factorial
    """
    def __init__(self, num):
        self.num = num

    def calculate(self):
        """
        this function calculates the factorial of the number

        param: num is the number wanted to calculate its factorial,
        type: integer

        return: factorial of the number,
        rtype: integer
        """
        if self.num == 0:
            return 1
        if self.num == 1:
            return 1
        
        return self.num * Factorial(self.num - 1).calculate()


f_1 = Factorial(5)
print(f_1.calculate())
```

```python
class SumEven:
    """
    class SumEven: handles the operation of calculating the sum of even numbers in a list

    Attributes:
            numbers(list): list of integers to calculate the sum of even numbers from
    """
    def __init__(self, numbers):
        self.numbers = numbers

    def calculate(self):
        """
        this function calculates the sum of even numbers in the list

        param: numbers is the list of integers,
        type: list

        return: sum of even numbers,
        rtype: integer
        """
        total = 0
        for i in self.numbers:
            if i % 2 == 0:
                total += i
        return total


nums = [1, 2, 3, 4, 5, 6]
s1 = SumEven(nums)
print(s1.calculate())
```

```python
class ReverseList:
    """
    class ReverseList: handles the operation of reversing a list provided by the user

    Attributes:
            lst(list): list of elements entered by the user
    """
    def __init__(self):
        self.lst = input("Enter your list: ").split()

    def reverse(self):
        """
        this function takes a list from user and returns its reverse

        return: reversed list,
        rtype: list
        """
        reversed_lst = self.lst[::-1]
        return reversed_lst


r1 = ReverseList()
print(r1.reverse())
```

```python
class Median:
    """
    class Median: handles the operation of calculating the median of a list of numbers

    Attributes:
            lst(list): list of numbers to calculate the median from
    """
    def __init__(self, lst: list):
        self.lst = lst

    def med(self):
        """
        this function takes a list of numbers and returns the median of these numbers

        param: lst is the list of numbers,
        type: list

        return: median of the numbers,
        rtype: float if length of list is odd, int otherwise
        """
        self.lst.sort()
        length = len(self.lst)
        if length % 2 == 0:
            mid = length // 2
            med = (self.lst[mid] + self.lst[mid - 1]) / 2
        else:
            med = self.lst[length // 2]
        return med


lst = [5, 2, 7, 3, 1]
m1 = Median(lst)
print(m1.med())
```

```python
class Mode:
    """
    class Mode: handles the operation of calculating the mode of a list of numbers

    Attributes:
            lst(list): list of numbers to calculate the mode from
    """
    def __init__(self, lst: list):
        self.lst = lst

    def mod(self):
        """
        this function takes a list of numbers and calculates its mode

        param: lst is the list of numbers,
        type: list

        return: mode of the numbers,
        rtype: int
        """
        count_lst = []
        for i in range(len(self.lst)):
            n = self.lst.count(self.lst[i])
            count_lst.insert(i, n)
        biggest = max(count_lst)
        index = count_lst.index(biggest)
        mode = self.lst[index]
        return mode


lst = [5, 2, 7, 3, 2, 1, 5]
m_1 = Mode(lst)
print(m_1.mod())
```
