```python
def factorial(num):
    """
    """
    if num ==  0:
        return 1
    if num == 1:
        return 1
    return num * factorial(num-1) 
```

```python
factorial(5)
```

```python
def is_prime(num):
    """
    """
    for i in range(2, num):
        if num%i==0:
            return False
        else:
            return True
```

```python
is_prime(3)
```

```python
def reverse(words:str):
    """
    """
    split_words=words.split(" ")
    new_list=split_words[::-1]
    result=" ".join(new_list)
    return result
```

```python
reverse("hello world")
```
