# find the factorial of a number 
## Example
(factorial of 5) = 5 * 4 * 3 * 2 * 1 = 120

```python
# Write your answer here
def fact(num : int):
    """
    this function clculates the factorial of the number

    param: number is wanted to calculate its factorial,
    type: integer

    return: factorial,
    rtype: integer
    """
    if num==0:
        return 1
    if num==1:
        return 1
    
    return num*fact(num-1)
fact(5)
```

# Create a function that Calculates the Sum of Even Numbers in a List
## Example
numbers: [1, 2, 3, 4, 5, 6]

Output: 12

```python
# Write your answer here
def sum_even(numbers: list):
    """
    this function calculates the sum of even numbers in a list

    param: numbers is the list of numbers,
    type: list of integers

    return: sum of even numbers,
    rtype: integer
    """
    total = 0
    for i in numbers:
        if i % 2 == 0:
            total += i
    return total

numbers = [1, 2, 3, 4, 5, 6]
print(sum_even(numbers))
```

# create a function that calculates the median of a list of numbers (a median is the middle element in sorted list of numbers)

## Example
numbers: [5, 2, 7, 3, 1]

Output: 3

```python
# Write your answer here
def med(lst:list):
    """
    this function take a list of numbrs and return the median of these numbers

    param: list of numbers
    type: list

    return: median 
    rtype: float if  len of list odd, int O.W
    """

    lst.sort()
    len_=len(lst)
    if len_%2==0:
        len_=len_//2
        med=(lst[len_]+lst[len_-1])/2
    else:
        med=lst[len_//2]
    return med

lst=[5,2,7,3,1]
med(lst)
```

# create a function that calculates the mode of a list of number (mode is the most frequent element in a list)

## Example
lst: [5, 2, 7, 3, 2, 1, 5]

Output: 5

```python
# Write your answer here
def mod(lst):
    """
    this function takes a list of number and calculates its mode

    pram: list of numbers 
    type: list

    return: mode 
    rtype: int

    """
    count_lst=[]
    for i in range(len(lst)):
      n=lst.count(lst[i])
      count_lst.insert(i,n)  
    biggest=max(count_lst)
    index=count_lst.index(biggest)
    mod=lst[index]
    return mod
lst=[5,2,7,3,2,1,5]
mod(lst)

```

# Task: take the mark from the user and convert it to grade according to the following guidelines
1. 85-100 A
2. 75-85 B
3. 65-75 C
4. 50-65 D

```python
# Write your answer here
try:
    mark=int(input("Enter your mark between 0 and 100"))
    if  mark<0 or mark>100:
        print("invalid range of marks, mark should betwwen 0 and 100")
    elif mark>85:
        print("Your grade is A")
    elif mark>75:
        print("Your grade is B")
    elif mark>65:
        print("Your grade is C")
    elif mark>=50:
        print("Your grade is D")
    else:
        print("Your grade is F (failed to pass)")
except Exception as e:
    print(e)
```
