```python
def user_info(name, age):
    """
    this function take name and age from user and return again
    
    :parm_1 : take name of user from input interface
    type param_1 : str
    :parm_2 : take age of user from input interface
    type param_2 : int

    return : this function will print the name and age
    rtype : str, int
    """
    
    return f"my name is : {name}, and my age is : {age}"
```

```python
user_info("Ahmed", 22)
```

```python
user_info(22, "Ahmed") # positional argument function
```

```python
user_info(age = 22,name = "Ahmed")  # keyword argument function
```
