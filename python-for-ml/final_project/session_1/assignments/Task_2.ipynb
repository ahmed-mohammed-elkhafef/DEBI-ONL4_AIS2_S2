Task-1 
Suppose you have this email address “Amit_ml@gmail.edu” - Input Validation: Check if the input string contains exactly one "@" symbol and at least one "." after the "@" symbol. 
If it's not a valid email, return "Invalid email".
 - Extract Username: Extract and return the part of the email before the "@" 
symbol.
 - Extract Domain: Extract and return the domain (the part between "@" and the 
last "."). 
 - Check for Domain Ending: Check if the email ends with ".com". If it does, 
return "Commercial Domain". If it ends with ".edu", return "Educational Domain". 
Otherwise, return "Other Domain".

```python
input_email = input("Enter your email") # input email is: Amit_ml@gmail.edu
input_email=input_email.lower()
c= input_email.count("@")
email=""
domain=""
index=[]
if c ==1:
    s_email=input_email.split("@")
    if s_email[0] == "":
        print("Invalid email")
        exit()

    if "." in s_email[1] and not s_email[1].endswith("."):
        email=s_email[0]
        part=s_email[1]
        if ".." in part:
            print("Invalid email")
            exit()

        len1=len(s_email[1])
        for i in range(0,len1,1):
            if part[i] ==".":
                index.append(i)
        pass
        len2=len(index)
        for i in range(0,index[len2-1],1):
            domain+=part[i]
        if domain=="":
            print("Invalid email \"No domain name\"")
            exit()
        end="Commercial Domain" if input_email.endswith(".com") else ("Educational Domain" if input_email.endswith(".edu") else "Other Domain" )
        print(f"Username is: {email},\nDomain is: {domain},\nType of Domain: {end}")
    else:
        print("Invalid email")

else:
    print("Invalid email")
```

Task-2: 
Encoded Message: 
###!!@mocleW EPGTQ!!!6789 
Steps to Decode: 
1. Extract the core part of the message: "mocleW EPGTQ". 
2. Reverse the first word: "mocleW" becomes "Welcome". 
3. Replace shifted vowels in the second word: 
    o "EPGTQ": No vowels to change. 
4. Final decoded message: "Welcome PGTQ". 

```python
e_m="###!!@mocleW EPGTQ!!!6789"
m=""
for i in e_m:
    if i.isalpha() or i==" ":
        m+=i
m=m.split()
m1=m[0]
m2=m[1]
m3=""
for i in range(len(m1)-1,-1,-1):
    m3+=m1[i]
print(f"Encoded message is: {m3+"e"+ " " +m2}")
```

Task-3: 
Encoded Message: 
&&&**$gnirtS PLIO!!@1234 
Steps to Decode: 
1. Extract the core part of the message: "gnirtS PLIO". 
2. Reverse the first word: "gnirtS" becomes "String". 
3. Replace shifted vowels in the second word: 
o "PLIO": Replace I->E and O->U to get "PLEU". 
4. Final decoded message: "String PLEU".

```python
e_m="&&&**$gnirtS PLIO!!@1234"
m=""
for i in e_m:
    if i.isalpha() or i==" ":
        m+=i
m=m.split()
m1=m[0]
m2=m[1]
m3=""
for i in range(len(m1)-1,-1,-1):
    m3+=m1[i]
m2=m2.replace("I","E").replace("O","U")
print(f"Encoded message is: {m3+ " " +m2}")
```
