```python
import numpy as np
```

```python
X = [50, 60, 70, 80, 90] 
Y = [150, 180, 210, 240, 270]  
# x represent feature and y represent target

x=np.array(X)
y=np.array(Y)
```

```python
from Linear_Regression_Class import *
```

```python
model = LinearRegression_Ahmed(num_iteration=100, alpha=0.001)
theta_1, theta_0 = model.fit(x, y)
print(f"theta 0 (bias): {theta_0}, theta 1 (wieght): {theta_1}")
```

```python
pred_price = model.Pred_y(70)
print(pred_price)
```

Normalization


```python
x_min, x_max = x.min(), x.max()
y_min, y_max = y.min(), y.max()

x_norm = (x - x_min) / (x_max - x_min)
y_norm = (y - y_min) / (y_max - y_min)
```

```python
model = LinearRegression_Ahmed(num_iteration=100, alpha=0.001)
theta_1, theta_0 = model.fit(x_norm, y_norm)
print(f"theta 0 (bias): {theta_0}, theta 1 (weight): {theta_1}")
```

```python
pred_norm = (70 - x_min) / (x_max - x_min)
pred_price_norm = model.Pred_y(pred_norm)

pred_real = pred_price_norm * (y_max - y_min) + y_min
print("Predicted price for 70 m²:", pred_real)
```
