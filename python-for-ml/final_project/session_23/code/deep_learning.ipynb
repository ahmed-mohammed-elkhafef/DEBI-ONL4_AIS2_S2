```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def update_weights(weights, x, z, y, lr):
    m = x.shape[0]
    dw = 1/m*np.dot(x.T,(z-y))
    weights = weights -lr* dw
    return weights
```

```python
def update_bias(bias, x, z, y, lr):
    m = x.shape[0]
    db = (1/m)*np.sum(z-y)
    bias = bias - lr *db
    return bias
```

```python
def perceptron(x, weights, bias):
    z = np.dot(x, weights)+bias
    y_pred = sigmoid(z)
    y_p = [1 if y_pred > 0.5 else 0 for y in y_pred]
    return y_p
```

```python
def accuracy(y_pred, y_test):
    acuuracy = np.sum(y_pred == y_test)/len(y_test)
    return acuuracy
```
